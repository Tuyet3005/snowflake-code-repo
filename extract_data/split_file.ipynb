```python
import pandas as pd
```

```python
df = pd.read_csv("Superstore.csv",  encoding='latin1')
print(df.columns)
df.head(2)
```

```python
# customer

customer_df = df[['Customer ID', 'Customer Name']]
customer_df = customer_df.drop_duplicates()
print(customer_df.columns)
print(customer_df.shape)
customer_df.to_csv("customer.csv", index=False)
```

```python
# order_infor

order_infor_df = df[['Order ID', 'Order Date', 'Customer ID']]
order_infor_df = order_infor_df.drop_duplicates()
print(order_infor_df.columns)
print(order_infor_df.shape)
order_infor_df.to_csv("order_infor.csv", index=False)
order_infor_df.head(2)
```

```python
# order_product

order_product_df = df[['Order ID', 'Product ID', 'Sales', 'Quantity', 'Discount', 'Profit']]
order_product_df = order_product_df.drop_duplicates()
order_product_df.index.name = 'ID'
print(order_product_df.columns)
print(order_product_df.shape)
order_product_df.to_csv("order_product.csv")
order_product_df.head(2)
```

```python
# product

product_df = df[['Product ID', 'Category', 'Sub-Category', 'Product Name']]
product_df = product_df.drop_duplicates()
print(product_df.columns)
print(product_df.shape)
product_df.to_csv("product.csv", index=False)
product_df.head(2)
```

```python
# shipment

shipment_df = df[['Order ID', 'Ship Date', 'Ship Mode', 'Segment', 'Country', 'City', 'State',
       'Postal Code', 'Region']]
shipment_df = shipment_df.drop_duplicates()
shipment_df.index.name = 'Shipment ID'
print(shipment_df.columns)
print(shipment_df.shape)
shipment_df.to_csv("shipment.csv")
shipment_df.head(2)
```
